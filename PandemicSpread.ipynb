```python
# combine deaths datasets to df_deaths
dfs_deaths = [aleppo_deaths_df, colombia_deaths_df, iran_deaths_df, karachi_deaths_df,\
       lebanon_deaths_df, nairobi_deaths_df, saudiarabia_deaths_df,\
       thailand_deaths_df, turkey_deaths_df, venezuela_deaths_df,\
       yemen_deaths_df]

df_deaths = pd.concat(dfs_deaths)
```

```python
# combine df and df_deaths
merged_df = pd.merge(df, df_deaths, on='ID')
merged_df = merged_df.drop(['USER_WARNING_x', 'USER_WARNING_y'], axis=1)
merged_df = merged_df[['ID', 'REGION', 'AGE', 'GENDER', 'DATE', 'DATE_OF_DEATH', 'SYNDROME']]
```

```python
# Calculate number of days lived
merged_df['DAYSLIVED'] = (merged_df['DATE_OF_DEATH'] - merged_df['DATE']).dt.days

# Add DAYSLIVED column to df_deaths
merged_df['DAYSLIVED'] = merged_df['DAYSLIVED']

merged_df.head()
```

##### Deceased Patients by Country

```python
# Filter out only deceased patients
deceased_df = merged_df.dropna(subset=['DATE_OF_DEATH'])
# Group by country and count the number of deceased patients
country_deaths = deceased_df.groupby('REGION')['ID'].count()

# Create a dataframe with the counts and country information
country_data = pd.DataFrame({'REGION': country_deaths.index, 'DEATHS': country_deaths.values})

# Create a choropleth map with plotly
fig = go.Figure(data=go.Choropleth(
    locations=country_data['REGION'], # Set location names
    z=country_data['DEATHS'], # Set color values
    locationmode='country names', # Use country names as location mode
    colorscale='Reds', # Choose color scale
    autocolorscale=False,
    marker_line_color='white', # Set marker line color
    colorbar=dict(
        title='Number of Deceased Patients',
        thickness=20,
        ticklen=3,
        tickformat=',',
        ticks='outside'
    ),
    zmin = 0,
    zmax = 30000
))

# Add title and annotations to the figure
fig.update_layout(
    title='Dead Patients by Country',
    geo=dict(
        showframe=False, # Remove frame around map
        showcoastlines=True, # Show country borders
        projection_type='equirectangular'
    ),
    annotations=[dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Cities were replaced with corresponding country',
        showarrow=False
    )]
)

fig.show()
```
